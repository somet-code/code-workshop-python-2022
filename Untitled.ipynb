```python
import math

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
from metpy.calc import reduce_point_density
from metpy.units import units
import numpy as np
import pandas as pd
from scipy.interpolate import griddata
```

```python
current_url = 'https://www.mesonet.org/data/public/mesonet/current/current.csv.txt'
```

```python
df = pd.read_csv(current_url,
                 sep=',',
                 header=0,
                 usecols=['STID', 'LAT', 'LON', 'PRES'],
                 index_col='STID',
                 na_values='',
                )
df['PRES'] = pd.to_numeric(df['PRES'], errors='coerce')
df = df.dropna()
```

```python
xi = np.linspace(math.floor(df['LON'].min()), math.ceil(df['LON'].max()), len(df['LON']))
yi = np.linspace(math.floor(df['LAT'].min()), math.ceil(df['LAT'].max()), len(df['LAT']))
xi, yi = np.meshgrid(xi, yi);
```

```python
x, y, z = df['LON'], df['LAT'], df['PRES']
zi = griddata((x, y), z, (xi, yi))
```

```python
mask = np.random.choice([True, False], size=df['LON'].size)
```

```python
lons = df['LON'].mask(mask).dropna()
lats = df['LAT'].mask(mask).dropna()
pres = df['PRES'].mask(mask).dropna()
```

```python
fig, ax = plt.subplots(figsize=(10,8), subplot_kw={'projection':ccrs.LambertConformal()})

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.STATES)
ax.scatter(lons, lats, marker='o', transform=ccrs.PlateCarree())

transform = ccrs.PlateCarree()._as_mpl_transform(ax)

for i in range(len(pres)):
    ax.annotate(str(round(pres[i])), (lons[i], lats[i]), xycoords=transform)
```
